```python
import tkinter as tk
import random

# Function to determine the game result
def game_result(user_choice, computer_choice):
    if user_choice == computer_choice:
        return "It's a tie!"
    if (user_choice == 's' and computer_choice == 'w') or \
       (user_choice == 'w' and computer_choice == 'g') or \
       (user_choice == 'g' and computer_choice == 's'):
        return "You win!"
    else:
        return "You lose!"

# Functions to play the game for each choice
def play_snake():
    play_game('s')

def play_water():
    play_game('w')

def play_gun():
    play_game('g')

def play_game(choice):
    user_choice = choice
    computer_choice = random.choice(choices)
    result = game_result(user_choice, computer_choice)
    result_var.set(f"Computer chose: {computer_choice.upper()} \n{result}")
    highlight_result(result)

# Function to highlight result
def highlight_result(result):
    if "win" in result.lower():
        result_label.config(bg="lightgreen", fg="black")
    elif "lose" in result.lower():
        result_label.config(bg="lightcoral", fg="black")
    else:
        result_label.config(bg="yellow", fg="black")

# Set up the main window
root = tk.Tk()
root.title("Snake Water Gun Game")
root.geometry("400x400")
root.configure(bg='red')

# Choices for the game
choices = ['s', 'w', 'g']

# Result display
result_var = tk.StringVar()
result_label = tk.Label(root, textvariable=result_var, font=("Times New Roman", 18, "bold"), bg='red', fg='white')
result_label.pack(pady=20)

# Buttons for user choices
button_frame = tk.Frame(root, bg='yellow')
button_frame.pack()

snake_button = tk.Button(button_frame, text="Snake", command=play_snake, bg='lightgreen', fg='black', font=("Times New Roman", 14, "bold"), bd=5, relief=tk.RAISED)
snake_button.grid(row=0, column=0, padx=20, pady=10)

water_button = tk.Button(button_frame, text="Water", command=play_water, bg='lightblue', fg='black', font=("Times New Roman", 14, "bold"), bd=5, relief=tk.RAISED)
water_button.grid(row=0, column=1, padx=20, pady=10)

gun_button = tk.Button(button_frame, text="Gun", command=play_gun, bg='black', fg='white', font=("Times New Roman", 14, "bold"), bd=5, relief=tk.RAISED)
gun_button.grid(row=0, column=2, padx=20, pady=10)

# Add End Game button
end_button = tk.Button(root, text="End Game", command=root.destroy, bg='orange', fg='black', font=("Times New Roman", 14, "bold"), bd=5, relief=tk.RAISED)
end_button.pack(pady=20)

# Run the application
root.mainloop()
```
